```python

def peel_df(df):
    data = []
    for i in range(len(df)):
        d = {}
        for n,key in enumerate(list(df)[1:]):
            d[key] = df.iloc[i][n+1]
        data.append(d)
    return df
data  = peel_df(df)
```
